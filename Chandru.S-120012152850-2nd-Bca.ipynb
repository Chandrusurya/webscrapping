## scrapping website

```python
url='https://www.flipkart.com/search?q=teddy+bear&sid=tng%2Cclb%2Cxv3&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=teddy+bear%7CTeddy+Bears&requestId=75f3aa01-299f-4d39-bcc6-db4a1e561fb2&as-searchtext=teddy'
```

# website response

```python
res=requests.get(url)
```

```python
res
```

## parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

## Teddy Names

```python
Teddy=soup.select('.s1Q9rs')
```

```python
Teddy
```

```python
price=soup.select('._8VNy32 ._30jeq3')
```

```python
price
```

```python
Offer=soup.select('._8VNy32 span')
```

```python
Offer
```

```python
df=cs.DataFrame(Teddy,columns=["Teddy"])
```

```python
df
```

```python
df['price']=price
```

```python
df
```

```python
df['Offer']=Offer
```

```python
df
```

```python
df.to_csv('Tdetails.csv')
```
